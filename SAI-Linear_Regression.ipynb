```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/MargaritaNuridj/csula_data_science/master/spooky-author-identification/train.csv")
```

a- Check out the dataset

```python
df.head(20)
```

```python
df.author.value_counts()
```

```python
# change author to numbers
df['author'] = df['author'].str.replace('EAP', '0')
df['author'] = df['author'].str.replace('HPL', '1')
df['author'] = df['author'].str.replace('MWS', '2')
df.head(15)
```

```python
# filter the DataFrame using an OR condition to get only MWS and EAP (MWS is latter in ASKII order)
MWS_EAP = df[(df.author=='2') | (df.author=='0')]
```

```python
MWS_EAP.shape
MWS_EAP.head(15)
```

Define X and y from the new DataFrame, and then split X and y into training and testing sets, using the text as the only feature and the author as the response.

Keep in mind that X should be a pandas Series (not a DataFrame), since we will pass it to CountVectorizer in the task that foll

```python
# define X and y
X = MWS_EAP.text
y = MWS_EAP.author

X.shape

y.shape
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into testing and training:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, 
                                                    random_state=6)
```

```python
# examine the object shapes
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# import and instantiate CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
# fit and transform X_train into X_train_dtm
X_train_dtm = vect.fit_transform(X_train)
X_train_dtm.shape
```

```python
# transform X_test into X_test_dtm
X_test_dtm = vect.transform(X_test)
X_test_dtm.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X_train_dtm, y_train)
```

```python
def best_and_worst(nums):
    indexes = [0, 0]
    values = [nums[0], nums[0]]

    for i in range(len(nums)):
        if nums[i] > values[0]:
            indexes[0] = i
            values[0] = nums[i]

        if nums[i] < values[1]:
            indexes[1] = i
            values[1] = nums[i]

    return indexes

# Even large, negative coefficients affect the model's output
coefficients = [abs(coefficient) for coefficient in linear_regression.coef_]
print(coefficients)

# MWS, EAP
best, worst = best_and_worst(coefficients)
```

```python
from numpy import sqrt
from sklearn.metrics import mean_squared_error

y_predict = linear_regression.predict(X_test_dtm)
print(y_predict)
```

```python
# Calculating "Mean Square Error" (MSE):
# Using numpy sqrt function to take the square root and calculate "Root Mean Square Error" (RMSE)
print('RMSE: {:.3f}'.format(sqrt(mean_squared_error(y_test, y_predict))))

```

```python
from sklearn.model_selection import cross_val_score

mse_list = cross_val_score(linear_regression, X_test_dtm, y_test, cv=10, scoring='neg_mean_squared_error')


# Notice that "cross_val_score" by default provides "negative" values for "mse" to clarify that mse is error.
# in order to calculate root mean square error (rmse), we have to make them positive!
mse_list_positive = -mse_list

# using numpy sqrt function to calculate rmse:
rmse_list = np.sqrt(mse_list_positive)
print(rmse_list)
```

```python
# calculate the average RMSE as final result of cross validation:
print('Cross-Validation RMSE: {:.3f}'.format(rmse_list.mean()))
```
